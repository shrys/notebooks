generators, couroutines (sub routine with a state)
http://www.dabeaz.com/coroutines/

```python
def fibgen(N): # a coroutine
    a,b,n = 0,1,0
    while n < N:
        yield a
        a,b = b,a+b
        n += 1
```

```python
sum(fibgen(100))
```

```python
type(fibgen(100))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('fruits_weight_sphercity.csv', converters={"labels":lambda x : 0 if x.lower() == 'orange' else (1 if x.lower() == 'apple' else -1)})
```

```python
data[30:].head()
```

```python
data1 = data[0:60:3] # Considering only 20 samples
data1.head()
```

```python
len(data)
```

```python
len(data1)
```

```python
train = data1
print(type(train))
test = data[1:50:10]
print(test)
print(len(test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
neigh = KNeighborsClassifier(n_neighbors=k,metric='euclidean', algorithm='brute')
```

```python
traindata = train.values[:, 1:3] #extracting 1,2 columns
print(train.values)
print(traindata)
```

```python
neigh.fit(traindata,train.labels)
```

```python
test1 = test.values[:,1:3]
```

```python
pred1 = neigh.predict(test1)
```

```python
print(pred1,'predictions')
print(test.values[:,3],'labels')
```

```python
plt.style.use('default')
oranges = train[train.labels == 0]
apples = train[train.labels == 1]
plt.scatter(oranges.Sphericity, oranges.Weight,color='orange',marker='o') 
plt.scatter(apples.Sphericity, apples.Weight,color='red',marker='o')
oranges = test[test.labels == 0]
apples = test[test.labels == 1]
plt.scatter(oranges.Sphericity, oranges.Weight,color='orange',marker='*',s=200) 
plt.scatter(apples.Sphericity, apples.Weight,color='red',marker='*',s=200)
plt.xlabel('Sphericity')
plt.ylabel('Weight')
plt.legend(['Oranges','Apples'])
plt.title('train, test points before scaling')
```

```python
import numpy as np
pred2 = np.column_stack((test1,pred1.T))
print(pred2)
```

```python
plt.style.use('default')
oranges = train[train.labels == 0]
apples = train[train.labels == 1]
oranges1 = pred2[pred2[:,2]==0]
apples1 = pred2[pred2[:,2]==1]
plt.scatter(oranges.Sphericity, oranges.Weight,color='orange',marker='o')
plt.scatter(apples.Sphericity, apples.Weight,color='red',marker='o')
plt.scatter(oranges1[:,0],oranges1[:,1],color='orange',marker='*',s=200)
plt.scatter(apples1[:,0],apples1[:,1],color='red',marker='*',s=200)
plt.xlabel('Sphericity')
plt.ylabel('Weight')
plt.legend(['Oranges','Apples'])
plt.title('Predictions before scaling')
plt.show()
```
